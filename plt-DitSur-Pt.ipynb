# Plot Ditector surface 

### define the function

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def MakeDitSurColorMap(filename):
    test2 = pd.read_csv(filename)
    test2 = test2.drop(columns = "z")
    box = np.zeros((100,100))
    cell = 30.5 / 100
    for i in range(100):
        for j in range(100):

            test_1 = test2[test2["x"] >= i * cell -15.25]
            test_2 = test_1[test_1["x"] < (i+1) * cell -15.25]
            test_3 = test_2[test_2["y"] >= j * cell -15.25]
            test_4 = test_3[test_3["y"] < (j+1) * cell -15.25]

            box[i,j] = len(test_4)
    return box
```

```python
def MakeDitSurHist(filename):
    Points = pd.read_csv(filename)
        
    Sx = Points.drop(columns= ['y','z'])
    Sy = Points.drop(columns= ['x','z'])
    
    bins_list = np.zeros(200)
    for bin_counter in range(200):
        bins_list[bin_counter] = bin_counter * 0.15 -15
        
    return bins_list, Sx, Sy
```

### Import data 

```python
bins, histx_my_beam, histy_my_beam = MakeDitSurHist("/home/yoshijun/Work/tutorial01/TestBench/data/DitPts_broadbeam.csv")
```
